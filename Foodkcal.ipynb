### Problem description, target audience, MVP description

Salah satu permasalahan yang sering terjadi pada masyarakat modern adalah kurangnya informasi mengenai kandungan kalori dan komposisi makanan, sehingga dapat menyebabkan gizi buruk dan gangguan kesehatan. Layanan FoodScan memecahkan masalah ini dengan memberikan informasi rinci kepada pengguna tentang kandungan kalori makanan dan menyarankan resep berdasarkan pengenalan makanan dalam foto menggunakan deteksi objek jaringan saraf.

**The FoodScan service can satisfy the following user needs**:
* Perlu menggunakan produk secara efektif: Pengguna mungkin memiliki bahan-bahan yang terbatas dan ingin menemukan cara untuk menggunakannya secara paling efektif. "FoodScan" membantu mereka menemukan resep yang dapat dibuat dari bahan-bahan yang tersedia.
* Perlu menghemat waktu: Mencari resep secara online atau di buku bisa memakan waktu dan tenaga. "FoodScan" menawarkan cara mudah dan cepat untuk menemukan resep yang cocok dengan daftar bahan pengguna, menghemat waktu dan tenaga.
* Perlunya Mengurangi Limbah Makanan: Banyak orang yang menerapkan gaya hidup yang lebih ramah lingkungan dan ingin mengurangi dampak buruk terhadap lingkungan. FoodScan membantu mengurangi sampah makanan dengan memungkinkan pengguna menggunakan makanan yang sudah mereka miliki alih-alih membuangnya.
* Perlunya Pengendalian Berat Badan: Pengguna yang ingin mengontrol berat badannya dapat menggunakan penghitungan kalori untuk menentukan asupan kalori harian dan melacak berapa banyak kalori yang mereka konsumsi dalam makanan.
* Kebutuhan untuk Mencapai Tujuan Atletik: Atlet dan penggemar kebugaran dapat menggunakan penghitungan kalori untuk menentukan keseimbangan energi mereka dan mencapai tujuan atletik tertentu..

  **Target** produknya adalah orang-orang yang menjalani gaya hidup sehat, melakukan kebugaran atau mencoba mengontrol pola makannya. Ini juga termasuk orang-orang dengan alergi atau pembatasan diet yang perlu memantau dengan cermat bahan-bahan dan kandungan kalori makanan. Selain itu, produk ini akan bermanfaat bagi mereka yang sedang mencari resep baru dan menarik untuk mendiversifikasi pola makan dan menikmati makanan enak dan sehat..

**Deskripsi layanan:**
Layanan FoodScan didasarkan pada jaringan saraf yang memungkinkan Anda mengenali makanan dalam foto dan menentukan kandungan kalorinya. Pengguna akan dapat mengambil foto makan siang, makan malam, atau makanan lainnya dan menerima informasi rinci tentang kalori dalam produk tersebut.
Selain itu, layanan ini akan menawarkan kepada pengguna daftar resep yang dapat dibuat dari bahan-bahan yang dikenali. Ini akan membantu pengguna mendiversifikasi pola makan dan menyeimbangkan pola makan mereka.

### Mendefinisikan Kelas dan Konfigurasi Kumpulan Data

```python
class FoodChallengeDataset(utils.Dataset):
    def load_dataset(self, dataset_dir, load_small=False, return_coco=True):
        """ Loads dataset released for the AICrowd Food Challenge
            Params:
                - dataset_dir : root directory of the dataset (can point to the train/val folder)
                - load_small : Boolean value which signals if the annotations for all the images need to be loaded into the memory,
                               or if only a small subset of the same should be loaded into memory
        """
        self.load_small = load_small
        if self.load_small:
            annotation_path = os.path.join(dataset_dir, "annotation-small.json")
        else:
            annotation_path = os.path.join(dataset_dir, "annotations.json")

        image_dir = os.path.join(dataset_dir, "images")
        print("Annotation Path ", annotation_path)
        print("Image Dir ", image_dir)
        assert os.path.exists(annotation_path) and os.path.exists(image_dir)

        self.coco = COCO(annotation_path)
        self.image_dir = image_dir

        # Load all classes (Only Building in this version)
        classIds = self.coco.getCatIds()

        # Load all images
        image_ids = list(self.coco.imgs.keys())

        # register classes
        for _class_id in classIds:
            self.add_class("crowdai-food-challenge", _class_id, self.coco.loadCats(_class_id)[0]["name"])

        # Register Images
        for _img_id in image_ids:
            assert(os.path.exists(os.path.join(image_dir, self.coco.imgs[_img_id]['file_name'])))
            self.add_image(
                "crowdai-food-challenge", image_id=_img_id,
                path=os.path.join(image_dir, self.coco.imgs[_img_id]['file_name']),
                width=self.coco.imgs[_img_id]["width"],
                height=self.coco.imgs[_img_id]["height"],
                annotations=self.coco.loadAnns(self.coco.getAnnIds(
                                            imgIds=[_img_id],
                                            catIds=classIds,
                                            iscrowd=None)))

        if return_coco:
            return self.coco

    def load_mask(self, image_id):
        """ Loads instance mask for a given image
              This function converts mask from the coco format to a
              a bitmap [height, width, instance]
            Params:
                - image_id : reference id for a given image

            Returns:
                masks : A bool array of shape [height, width, instances] with
                    one mask per instance
                class_ids : a 1D array of classIds of the corresponding instance masks
                    (In this version of the challenge it will be of shape [instances] and always be filled with the class-id of the "Building" class.)
        """

        image_info = self.image_info[image_id]
        assert image_info["source"] == "crowdai-food-challenge"

        instance_masks = []
        class_ids = []
        annotations = self.image_info[image_id]["annotations"]
        # Build mask of shape [height, width, instance_count] and list
        # of class IDs that correspond to each channel of the mask.
        for annotation in annotations:
            class_id = self.map_source_class_id(
                "crowdai-food-challenge.{}".format(annotation['category_id']))
            if class_id:
                m = self.annToMask(annotation,  image_info["height"],
                                                image_info["width"])
                # Some objects are so small that they're less than 1 pixel area
                # and end up rounded out. Skip those objects.
                if m.max() < 1:
                    continue

                # Ignore the notion of "is_crowd" as specified in the coco format
                # as we donot have the said annotation in the current version of the dataset

                instance_masks.append(m)
                class_ids.append(class_id)
        # Pack instance masks into an array
        if class_ids:
            mask = np.stack(instance_masks, axis=2)
            class_ids = np.array(class_ids, dtype=np.int32)
            return mask, class_ids
        else:
            # Call super class to return an empty mask
            return super(FoodChallengeDataset, self).load_mask(image_id)


    def image_reference(self, image_id):
        """Return a reference for a particular image

            Ideally you this function is supposed to return a URL
            but in this case, we will simply return the image_id
        """
        return "crowdai-food-challenge::{}".format(image_id)
    # The following two functions are from pycocotools with a few changes.

    def annToRLE(self, ann, height, width):
        """
        Convert annotation which can be polygons, uncompressed RLE to RLE.
        :return: binary mask (numpy 2D array)
        """
        segm = ann['segmentation']
        if isinstance(segm, list):
            # polygon -- a single object might consist of multiple parts
            # we merge all parts into one mask rle code
            rles = maskUtils.frPyObjects(segm, height, width)
            rle = maskUtils.merge(rles)
        elif isinstance(segm['counts'], list):
            # uncompressed RLE
            rle = maskUtils.frPyObjects(segm, height, width)
        else:
            # rle
            rle = ann['segmentation']
        return rle

    def annToMask(self, ann, height, width):
        """
        Convert annotation which can be polygons, uncompressed RLE, or RLE to binary mask.
        :return: binary mask (numpy 2D array)
        """
        rle = self.annToRLE(ann, height, width)
        m = maskUtils.decode(rle)
        return m
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers as KL
from tensorflow.keras import backend as K

class NormalizeBoxes(KL.Layer):
    def __init__(self):
        super(NormalizeBoxes, self).__init__()

    def call(self, inputs):
        boxes, shape = inputs
        h, w = shape[0], shape[1]
        return boxes / tf.convert_to_tensor([h, w, h, w], dtype=tf.float32)

class MaskRCNN:
    def __init__(self, config):  # Accepting config as an argument
        self.config = config

    def build(self, mode):  # No need to add config as a parameter
        input_gt_boxes = KL.Input(shape=(None, 4), name="input_gt_boxes", dtype=tf.float32)
        input_image = KL.Input(shape=self.config['IMAGE_SHAPE'], name="input_image")  # Using self.config

        # Create an instance of the custom NormalizeBoxes layer
        normalize_boxes_layer = NormalizeBoxes()

        # Normalize ground truth boxes using the custom layer
        gt_boxes = normalize_boxes_layer([input_gt_boxes, tf.shape(input_image)[1:3]])

        # Continue building the model with other layers...
        # Example: a simple backbone (e.g., ResNet or similar)
        backbone = self.build_backbone(input_image)

        # Example: a Region Proposal Network (RPN)
        rpn_class, rpn_bbox = self.rpn(backbone)

        # Other model components (classification head, mask head, etc.)
        mrcnn_class, mrcnn_bbox, mrcnn_mask = self.build_heads(backbone)

        # Create the model
        model = keras.Model(inputs=[input_image, input_gt_boxes],
                            outputs=[gt_boxes, rpn_class, rpn_bbox, mrcnn_class, mrcnn_bbox, mrcnn_mask],
                            name="mask_rcnn_model")
        return model

    def build_backbone(self, input_image):
        # Define a simple backbone network (example: ResNet)
        # Replace with actual backbone architecture
        x = KL.Conv2D(64, (3, 3), padding='same')(input_image)
        x = KL.MaxPooling2D(pool_size=(2, 2))(x)
        return x

    def rpn(self, backbone_output):
        # Example RPN definition
        rpn_class = KL.Conv2D(9, (1, 1), activation='sigmoid', name='rpn_class')(backbone_output)
        rpn_bbox = KL.Conv2D(9 * 4, (1, 1), name='rpn_bbox')(backbone_output)
        return rpn_class, rpn_bbox

    def build_heads(self, backbone_output):
        # Example head for classification and bounding box regression
        mrcnn_class = KL.Conv2D(self.config['NUM_CLASSES'], (1, 1), activation='softmax', name='mrcnn_class')(backbone_output)
        mrcnn_bbox = KL.Conv2D(self.config['NUM_CLASSES'] * 4, (1, 1), name='mrcnn_bbox')(backbone_output)
        mrcnn_mask = KL.Conv2D(self.config['NUM_CLASSES'], (1, 1), activation='sigmoid', name='mrcnn_mask')(backbone_output)
        return mrcnn_class, mrcnn_bbox, mrcnn_mask

# Usage example:
config = {
    'IMAGE_SHAPE': (None, None, 3),  # Replace with actual shape
    'NUM_CLASSES': 80  # Replace with actual number of classes
}

# Initialize your model
mask_rcnn_model = MaskRCNN(config)
model = mask_rcnn_model.build(mode="training")  # Only mode is needed
model.summary()  # Optional: To see the model summary
```

### Fungsi penghitungan kalori

```python
import pandas as pd

def find_food_calories(food_list):
    # Load the calories dataset
    data = pd.read_csv('/content/input/food-and-their-calories/kalori.csv')

    # Standardize column names for easier reference
    data.rename(columns={'name': 'Food', 'calories': 'Calories', 'proteins': 'Protein', 'fat': 'Fat', 'carbohydrate': 'Carbs'}, inplace=True)

    total_calories = 0
    results = []

    # Process each food item in the provided list
    for food in food_list:
        # Filter the data to find entries that match the food item
        filtered_data = data[data['Food'].str.contains(food, case=False, na=False)]

        if not filtered_data.empty:
            # Extract nutrient data from the first matched result
            protein = filtered_data['Protein'].iloc[0] if 'Protein' in filtered_data else 0
            fat = filtered_data['Fat'].iloc[0] if 'Fat' in filtered_data else 0
            carbs = filtered_data['Carbs'].iloc[0] if 'Carbs' in filtered_data else 0
            calories_str = filtered_data['Calories'].iloc[0]

            # Convert calories to an integer, handling different formats
            if isinstance(calories_str, float):
                calories = int(calories_str)
            else:
                try:
                    calories = int(calories_str.split(' ')[0])  # Extract the number before 'cal'
                except (ValueError, IndexError):
                    calories = 0

            # Add to total calories
            total_calories += calories

            # Create result text for the current food item
            result_text = (
                f"Food: {filtered_data['Food'].iloc[0]}\n"
                f"Protein: {protein if pd.notna(protein) else 0} g\n"
                f"Lemak: {fat if pd.notna(fat) else 0} g\n"
                f"Karbohidrat: {carbs if pd.notna(carbs) else 0} g\n"
                f"Calories: {calories} cal"
            )
            results.append(result_text)

    # Combine all results into a single string
    result_text = '\n\n'.join(results)
    result_text += f"\n\nTotal Calories: {total_calories} cal"

    return result_text

# Example usage
food_list = ['Ayam', 'Apel', 'es krim', 'kopi','pisang mas']
result = find_food_calories(food_list)
print(result)


```

### Fungsi pencarian resep:

```python
import pandas as pd

class_names_indonesian = []

def search_recipes(ingredients_list):
    dat1 = pd.read_csv('/content/input/resep/dataset-ayam.csv')
    dat2 = pd.read_csv('/content/input/resep/dataset-sapi.csv')
    dat3 = pd.read_csv('/content/input/resep/dataset-ikan.csv')
    dat4 = pd.read_csv('/content/input/resep/dataset-tahu.csv')
    dat5 = pd.read_csv('/content/input/resep/dataset-telur.csv')
    dat6 = pd.read_csv('/content/input/resep/dataset-udang.csv')
    dat7 = pd.read_csv('/content/input/resep/dataset-kambing.csv')
    dat8 = pd.read_csv('/content/input/resep/dataset-tempe.csv')
    data = pd.concat([dat1, dat2, dat3, dat4, dat5, dat6, dat7, dat8], ignore_index=True)


    data['Ingredients'] = data['Ingredients'].fillna('')

    ingredients_list_lower = [ingredient.lower() for ingredient in ingredients_list]

    filtered_data = data[data['Ingredients'].apply(lambda x: all(ingredient.lower() in x.lower() for ingredient in ingredients_list_lower))]

    filtered_data = filtered_data.head(3)

    recipes_text = ""
    for index, row in filtered_data.iterrows():
        recipes_text += f"Resep #{index+1}:\n"
        recipes_text += f"URL: {row['URL']}\n"
        recipes_text += f"Judul: {row['Title']}\n"
        recipes_text += f"Bahan-bahan: {row['Ingredients']}\n\n"

    return recipes_text


ingredients_list = class_names_indonesian
results = search_recipes(ingredients_list)

print(results)
```
